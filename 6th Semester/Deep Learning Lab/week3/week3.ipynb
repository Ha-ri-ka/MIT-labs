### QUESTION 1

```python
import torch
import matplotlib.pyplot as plt
x=torch.tensor([12.4,14.3,14.5,14.9,16.1,16.9,16.5,15.4,17.0,17.9,18.8,20.3,22.4,19.4,15.5,16.7,17.3,18.4,19.2,17.4,19.5,19.7,21.2])
y=torch.tensor([11.2,12.5,12.7,13.1,14.1,14.8,14.4,13.4,14.9,15.6,16.4,17.7,19.6,16.9,14.0,14.6,15.1,16.1,16.8,15.2,17.0,17.2,18.6])
b=torch.rand([1],requires_grad=True)
w=torch.rand([1],requires_grad=True)
alpha=torch.tensor(0.001)
print(f"The inital params are:\nw={w.item()}\nb={b.item()}\nlearning rate={alpha}")
loss_list=[]
epochCount=100
for epoch in range(epochCount):
    loss=0.0
    for j in range(len(x)):
        y_p=w*x[j]+b
        error=y_p-y[j]
        loss+=(error)**2
    loss=loss/len(x)
    loss_list.append(loss.item())
    loss.backward()
    with torch.no_grad():
        w-=alpha*w.grad
        b-=alpha*b.grad
    w.grad.zero_()
    b.grad.zero_()
    if(epoch%10==0 or epoch==99):
        print(f"epoch {epoch}: w={w.item()} b={b.item()} loss={loss.item()}")
plt.plot(loss_list)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Epochs Vs Loss")
```

### QUESTION 2

```python
import torch
x=torch.tensor([2,4])
y=torch.tensor([20,40])
w=torch.tensor([1.0],requires_grad=True)
b=torch.tensor([1.0],requires_grad=True)
alpha=torch.tensor(0.001)
print(f"The inital params are:\nw={w.item()}\nb={b.item()}\nlearning rate={alpha.item()}")
epochNum=2
for epoch in range(epochNum):
    loss=0.0
    for j in range(len(x)):
        y_p=w*x[j]+b
        error=y_p-y[j]
        loss+=(error)**2
    loss=loss/len(x)
    loss.backward()
    with torch.no_grad():
        w-=alpha*w.grad
        b-=alpha*b.grad
    w.grad.zero_()
    b.grad.zero_()
print(f"At the end of {epochNum} epochs:\nw={w.item()}\nb={b.item()}\nloss={loss.item()}")
print("\nPredictions by the model:")
y_pred=[w*xval+b for xval in x]
for it in y_pred:
    print(it.item(),end=' ')
```
