Suppose you have a gold/silver price dataset with a single independent variable (X) and a dependent variable 
(Y).  You  want  to  fit  a  polynomial  regression  model  to  this  data.  Implement  the  process  of  selecting  the 
appropriate degree for the polynomial (e.g., linear, quadratic, cubic) based on the dataset using Python.

```python
import pandas as pd
data=pd.read_csv("gold.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(data['YEAR'],data['GOLD RATE'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import numpy as np
def polyRegression(degree,X,Y):
    x=np.array(X).reshape(-1,1)
    y=np.array(Y)
    polyFeatures=PolynomialFeatures(degree)
    x_poly=polyFeatures.fit_transform(x)
    model=LinearRegression()
    model.fit(x_poly,y)
    y_new=model.predict(x_poly)
    return y_new
```

```python
degree2_predictions=polyRegression(2,data['YEAR'],data['GOLD RATE'])
degree3_predictions=polyRegression(3,data['YEAR'],data['GOLD RATE'])
degree4_predictions=polyRegression(4,data['YEAR'],data['GOLD RATE'])
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(data['YEAR'],data['GOLD RATE'],color='green')
plt.plot(data['YEAR'],degree2_predictions,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 2')

plt.subplot(1,3,2)
plt.plot(data['YEAR'],data['GOLD RATE'],color='green')
plt.plot(data['YEAR'],degree3_predictions,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 3')

plt.subplot(1,3,3)
plt.plot(data['YEAR'],data['GOLD RATE'],color='green')
plt.plot(data['YEAR'],degree4_predictions,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 4')
plt.tight_layout()
plt.show()
```

Imagine  you  have  a  gold  and  silver  price  dataset  with  two  independent  variables  (X1  and  X2)  and  a 
dependent  variable  (Y).  Implement  in  python,  how  you  can  perform  multivariate  polynomial  regression  to 
model the relationship between the independent variables and the dependent variable.

```python
#let year and gold be independent, silver be dependent variable
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
def multipolyRegression(degree,X,Y):
    x=np.array(X)
    y=np.array(Y)
    polyFeatures=PolynomialFeatures(degree)
    x_poly=polyFeatures.fit_transform(x)
    model=LinearRegression()
    model.fit(x_poly,y)
    y_pred=model.predict(x_poly)
    return y_pred   
```

```python
import pandas as pd
data=pd.read_csv("silver.csv")
data.head()
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
degree2=multipolyRegression(2,x,y)
degree3=multipolyRegression(3,x,y)
degree4=multipolyRegression(4,x,y)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(data['YEAR'],data['SILVER RATE'],color='green')
plt.plot(data['YEAR'],degree2,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 2')

plt.subplot(1,3,2)
plt.plot(data['YEAR'],data['SILVER RATE'],color='green')
plt.plot(data['YEAR'],degree3,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 3')

plt.subplot(1,3,3)
plt.plot(data['YEAR'],data['SILVER RATE'],color='green')
plt.plot(data['YEAR'],degree4,color='black')
plt.legend(['actual','predicted'])
plt.title('degree 4')
plt.tight_layout()
plt.show()
```
