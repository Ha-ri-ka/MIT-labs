Q2. Suppose you have a gold/silver price dataset with a single independent variable (X) and a dependent variable 
(Y). You want to fit a logistic regression model to this data. Develop an example code snippet in Python. 

```python
import pandas as pd
data=pd.read_csv("gold.csv")
data.head()
```

```python
#classify gold prices to be high or low. if price > 7000, its high (1), else low(0).
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
data['class']=None
for i in range (len(data['YEAR'])):
    if data.iloc[i,-2]>=7000:
        data.iloc[i,-1]='high'
    else:
        data.iloc[i,-1]='low'
x=np.array(data['YEAR']).reshape(-1,1)
y=data['class']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
model=LogisticRegression(solver='liblinear')
model.fit(x_train,y_train)
y_pred=model.predict(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(x_test,y_test)
plt.xlabel('year'), plt.ylabel('gold rate')
plt.title('actual values')
plt.subplot(1,2,2)
plt.plot(x_test,y_pred)
plt.xlabel('year'), plt.ylabel('gold rate')
plt.title('predicted values')
plt.show()
```

Imagine  you  have  a  gold  and  silver  price    dataset  with  two  independent  variables  (X1  and  X2)  and  a 
dependent  variable  (Y).  Implement  in  python,  how  you  can  perform  the  logistic  regression  to  model  the 
relationship between the independent variables and the dependent variable

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
data = pd.read_csv("silver.csv")
data['class']=None
for i in range (len(data['YEAR'])):
    if data.iloc[i,-2]>=8000:
        data.iloc[i,-1]='high'
    else:
        data.iloc[i,-1]='low'
x = data.iloc[:,:2]
y = data.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
model = LogisticRegression(solver='liblinear')
model.fit(x_train, y_train)
y_pred = model.predict(x)
# Evaluate the model
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(data['YEAR'],y)
plt.xlabel('year'), plt.ylabel('silver rate')
plt.title('actual values')
plt.subplot(1,2,2)
plt.plot(data['YEAR'],y_pred)
plt.xlabel('year'), plt.ylabel('silver rate')
plt.title('predicted values')
plt.show()
```
