```python
import pandas as pd
import numpy as np
path="data.csv"
data=pd.read_csv(path)
data
```

```python
x=np.array(data['x'])
y=np.array(data['y'])
b0,b1=0,0
slopes,coefficeints,errors=[],[],[]
n=data.shape[0]
epochs=4
alpha=0.01
for epoch in range(epochs):
        for i in range(n):
                ypred=b0+b1*x[i]
                err=ypred-y[i]
                b0=b0-alpha*err
                b1=b1-alpha*err*x[i]
                slopes.append(b0)
                coefficeints.append(b1)
                errors.append(err)
                print(f'{i} x={x[i]} y={y[i]} b0={b0}  b1={b1}, error={err} ')
```

```python
print(f'after {epochs} epochs, b0={b0} and b1={b1}, error={err}')
print(f'Y={b0} + {b1} x')
ynew=b0 + b1 * x
errorList=ynew-y
```

```python
import matplotlib.pyplot as plt
plt.plot(errors)
plt.xlabel('iteration')
plt.ylabel('errors')
plt.title('variation of errors for each iteration')
plt.show()
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,ynew)
plt.xlabel('x')
plt.ylabel('y')
plt.title('predicted Vs actual')
plt.legend(['actual','predicted'])
plt.show()
```

```python
errSq=np.array([pow(err,2) for err in errorList])
MSE=(errSq.sum())/n
RMSE=pow(MSE,0.5)
MSE,RMSE
```
