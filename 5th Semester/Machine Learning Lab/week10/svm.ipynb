```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
cancer=load_breast_cancer()
x=cancer.data[:,:2]
y=cancer.target
svm=SVC(kernel="rbf",gamma=0.5,C=1.0)
svm.fit(x,y)
DecisionBoundaryDisplay.from_estimator( 
        svm, 
        x, 
        response_method="predict", 
        cmap=plt.cm.Spectral, 
        alpha=0.8, 
        xlabel=cancer.feature_names[0], 
        ylabel=cancer.feature_names[1], 
    ) 
  

plt.scatter(x[:, 0], x[:, 1],  
            c=y,  
            s=20, edgecolors="k") 
plt.show()
```

Use the IRIS dataset, implement the SVM classifier in python (make use of scikit-learn library), to do the 
following. 

a. Apply the kernel functions such as linear, polynomial,  Radial basis functions and Sigmoid. 

b.  Plot the scatter plot of the input features. 

c.  Plot the decision boundary.

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
def performSVM(krnl,x,y):
    if krnl=="rbf":    
        svm=SVC(kernel="rbf",gamma=0.5,C=1.0)
        svm.fit(x,y)
    elif krnl=="linear":
        svm=SVC(kernel="linear",C=1.0)
        svm.fit(x,y)
    elif krnl=="sigmoid":
        svm=SVC(kernel="sigmoid",gamma=0.5,C=1.0)
        svm.fit(x,y)
    elif krnl=="poly":
        svm=SVC(kernel="poly",C=1.0,degree=2)
        svm.fit(x,y)
    return(svm)    

def plotBoundary(svm,x,y,xlbl,ylbl,krnl):
    DecisionBoundaryDisplay.from_estimator( 
            svm, 
            x, 
            response_method="predict", 
            cmap=plt.cm.Spectral, 
            alpha=0.8, 
            xlabel=xlbl, 
            ylabel=ylbl,
        )
    plt.scatter(x[:, 0], x[:, 1],  
                c=y,  
                s=20, edgecolors="k")
    plt.title(f"Classification using {krnl}") 
```

```python
dataset=load_iris()
x=dataset.data[:,:2]
y=dataset.target
xlable=dataset.feature_names[0]
ylable=dataset.feature_names[1]

kernels=["rbf","linear","sigmoid","poly"]
results=[]
for k in kernels:
    res=performSVM(k,x,y)
    results.append(res)
for r,k in zip(results,kernels):
    plotBoundary(r,x,y,xlable,ylable,k)
```
